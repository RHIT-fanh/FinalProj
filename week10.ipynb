```python
import os
import pandas as pd

# 指定本地数据集路径
path= r"C:\Users\fanh\.cache\kagglehub\datasets\wyattowalsh\basketball\versions\231"
csv_folder = os.path.join(path, "csv")

csv_files = [f for f in os.listdir(csv_folder) if f.endswith('.csv')]

csv_data = {file: pd.read_csv(os.path.join(csv_folder, file)) for file in csv_files}

for name, df in csv_data.items():
    print(f"{name} has {df.shape[1]} features:")
    print(df.columns.tolist())
    print("-" * 60)

```

```python
# 获取 inactive_players.csv 数据
inactive_players_df = csv_data['inactive_players.csv']

# 确保 game_df 定义
game_df = csv_data['game.csv']

# 计算每场比赛主队和客队的伤病人数
injury_counts = inactive_players_df.groupby(['game_id', 'team_id']).size().reset_index(name='injury_count')

# 将主队和客队的伤病人数分别合并到 game_df
home_injuries = injury_counts.rename(columns={'team_id': 'team_id_home', 'injury_count': 'home_injury_count'})
away_injuries = injury_counts.rename(columns={'team_id': 'team_id_away', 'injury_count': 'away_injury_count'})

# 合并主队和客队的伤病数据
game_injury_stats = game_df.merge(home_injuries, on=['game_id', 'team_id_home'], how='left') \
                           .merge(away_injuries, on=['game_id', 'team_id_away'], how='left')

# 填充缺失值为 0
game_injury_stats['home_injury_count'] = game_injury_stats['home_injury_count'].fillna(0).astype(int)
game_injury_stats['away_injury_count'] = game_injury_stats['away_injury_count'].fillna(0).astype(int)

# 提取最终的伤病统计数据
final_injury_stats = game_injury_stats[['game_id', 'home_injury_count', 'away_injury_count']]

print(final_injury_stats)
```

```python
# 读取 nba_cleaned_withoutAllStar.csv
nba_cleaned_withoutAllStar_df = pd.read_csv("nba_cleaned_withoutAllStar.csv")

# 合并 final_injury_stats 和 nba_cleaned_withoutAllStar_df
merged_df = nba_cleaned_withoutAllStar_df.merge(final_injury_stats, on='game_id', how='left')

# 查看合并后的数据
print(merged_df.head())

# 如果需要，可以将合并后的数据保存到新的 CSV 文件
merged_df.to_csv("merged_nba_cleaned.csv", index=False)
```

```python
# 添加 is_playoffs 特征
game_df['is_playoffs'] = game_df['season_type'] == 'Playoffs'

# 查看更新后的数据
print(game_df[['game_id', 'season_type', 'is_playoffs']].head())

```

```python
# 将 game_df 的 is_playoffs 列合并到 merged_df，避免重复列名
merged_df = merged_df.merge(game_df[['game_id', 'is_playoffs']], on='game_id', how='left', suffixes=('', '_game_df'))

# 查看更新后的数据
print(merged_df.head())
# 导出到 nba_22features.csv
merged_df.to_csv("nba_22features.csv", index=False)
```
